```python
# Step 1: Identify genes expressed in each group
# Create a boolean mask for genes expressed in Group 0 and Group 1
expressed_in_group0 = np.sum(B_cells[B_cells.obs['Groups'] == 'aHD'].X > 0, axis=0) > 5
expressed_in_group1 = np.sum(B_cells[B_cells.obs['Groups'] == 'aSLE'].X > 0, axis=0) > 5

# Step 2: Combine the masks to find genes expressed in both groups
expressed_in_both_groups = expressed_in_group0 & expressed_in_group1

# Step 3: Filter the AnnData object to keep only the genes expressed in both groups
B_cells_filtered = B_cells[:, expressed_in_both_groups]
```

```python
B_cells_filtered.shape
```

```python
# Step 1: Extract gene expression data for each group
group_HD = B_cells_filtered[B_cells_filtered.obs['Groups'].isin(['cHD', 'aHD'])]
group_SLE = B_cells_filtered[B_cells_filtered.obs['Groups'].isin(['cSLE', 'aSLE'])]

# Step 2: Get the gene expression matrices
expression_HD = group_HD.X  # Gene expression for cHD group
expression_SLE = group_SLE.X  # Gene expression for cSLE group

# Convert to DataFrame for easier manipulation (optional)
genes = B_cells_filtered.var_names  # Get gene names
df_expression_HD = pd.DataFrame(expression_HD.toarray(), columns=genes)
df_expression_SLE = pd.DataFrame(expression_SLE.toarray(), columns=genes)

# Step 3: Compare gene expression between groups
mean_expression_HD = df_expression_HD.mean()
mean_expression_SLE = df_expression_SLE.mean()
```

```python
X_group0 = df_expression_HD.values
X_group1 = df_expression_SLE.values
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.decomposition import PCA


# Step 1: Perform PCA with only 2 components for visualization
pca = PCA(n_components=15)
group0_pcs = pca.fit_transform(df_expression_HD)
group1_pcs = pca.fit_transform(df_expression_SLE)

# Separate into two groups based on 'Group'
# log_norm_data_2PC_group0 = log_norm_data[log_norm_data['Group'] == '0'][['PC1', 'PC2']]
# log_norm_data_2PC_group1 = log_norm_data[log_norm_data['Group'] == '1'][['PC1', 'PC2']]

# Step 3: Fit KDE using scipy.stats.gaussian_kde for the PCs of Group 0
kde_group0 = stats.gaussian_kde(group0_pcs.T, bw_method='scott')

# Step 4: Fit KDE using scipy.stats.gaussian_kde for the PCs of Group 1
kde_group1 = stats.gaussian_kde(group1_pcs.T, bw_method='scott')

# Evaluate densities for both groups using these points
density_group0 = kde_group0(group0_pcs.T)
density_group1 = kde_group1(group1_pcs.T)

# Step 6: Normalize densities to create mu and nu based on density estimates
# Perform min-max normalization
min_density_group0 = np.min(density_group0)
max_density_group0 = np.max(density_group0)
mu = (density_group0 - min_density_group0) / (max_density_group0 - min_density_group0)  # Scale between 0 and 1

min_density_group1 = np.min(density_group1)
max_density_group1 = np.max(density_group1)
nu = (density_group1 - min_density_group1) / (max_density_group1 - min_density_group1)  # Scale between 0 and 1
```

```python
# Compute the cost matrix between cells in the two groups
cost_matrix = ot.dist(X_group0, X_group1, metric='euclidean')
# cost_matrix /= cost_matrix.max()

min_cost_matrix = np.min(cost_matrix)
max_cost_matrix = np.max(cost_matrix)
cost_matrix_norm = (cost_matrix - min_cost_matrix) / (max_cost_matrix - min_cost_matrix)  # Scale between 0 and 1

# Define distributions over the cells in each group
# mu = np.ones(n_cells_group0) / n_cells_group0  # Uniform distribution for Group 0
# nu = np.ones(n_cells_group1) / n_cells_group1  # Uniform distribution for Group 1
```

```python
# Compute the OT plan
lambda_reg = 10  # Regularization parameter
transport_plan = ot.sinkhorn(mu, nu, cost_matrix_norm, lambda_reg)
```

```python
def verify_row_sums(transport_plan_normalized, tolerance=1e-6):
    """
    Verifies that each row of the normalized transport plan sums to 1 within a specified tolerance.

    Parameters:
    - transport_plan_normalized (np.ndarray): The normalized transport plan matrix.
    - tolerance (float): The numerical tolerance for checking row sums. Default is 1e-6.

    Returns:
    - None

    Outputs:
    - Prints a message indicating whether all rows sum to 1 within the tolerance.
    - If not, reports the number of rows that fail the check and their maximum deviation.
    """

    # Compute the row sums
    row_sums = np.sum(transport_plan_normalized, axis=1)

    # Check if all row sums are within the tolerance
    deviations = np.abs(row_sums - 1)
    max_deviation = np.max(deviations)
    failing_rows = np.where(deviations > tolerance)[0]
    num_failing = len(failing_rows)

    if num_failing == 0:
        print(f"All {transport_plan_normalized.shape[0]} rows sum to 1 within tolerance {tolerance}.")
    else:
        print(f"{num_failing} rows out of {transport_plan_normalized.shape[0]} do not sum to 1 within tolerance {tolerance}.")
        print(f"Maximum deviation from 1: {max_deviation}")
        print(f"Indices of failing rows: {failing_rows}")
```

```python
# Normalize the transport plan
row_sums = np.sum(transport_plan, axis=1, keepdims=True)
transport_plan_normalized = transport_plan / row_sums

# Compute the transported expression profiles for Group 0 cells
transported_expr = np.dot(transport_plan_normalized, X_group1)

# Compute the expression differences for each gene
expr_difference = transported_expr - X_group0
```

```python
# Compute the standard deviation of the differences for each gene
std_expr_difference = np.std(expr_difference, axis=0, ddof=1)

# Compute the standard error for each gene
n_cells = expr_difference.shape[0]
se_expr_difference = std_expr_difference / np.sqrt(n_cells)

# Adjust standard errors to avoid division by zero
se_expr_diff_adj = np.where(se_expr_difference == 0, np.finfo(float).eps, se_expr_difference)
```

```python
# Average the expression differences across cells
average_expr_diff = np.mean(expr_difference, axis=0)
```

```python
# Compute the Wald statistics
wald_stat = average_expr_diff / se_expr_diff_adj

# Compute two-tailed p-values
p_values = 2 * norm.sf(np.abs(wald_stat))

# Adjust p-values using Benjamini-Hochberg FDR correction
_, adj_p_values, _, _ = multipletests(p_values, method='fdr_bh')

# Verify the length of p_values and adj_p_values
print(f"Length of adj_p_values: {len(adj_p_values)}")    # Should be n_genes
```

```python
# Create the DE results DataFrame
genes = genes
de_results = pd.DataFrame({
    'Gene': genes,
    'Average_Expression_Diff': average_expr_diff,
    'Adj_SE': se_expr_diff_adj,
    'Wald_Statistic': wald_stat,
    'PValue': p_values,
    'AdjPValue': adj_p_values
})

# Sort the DataFrame by adjusted p-value
de_results.sort_values('AdjPValue', ascending=True, inplace=True)

# Display the top DE genes
print(de_results.head(20))
```

```python
# Define the significance threshold
significance_threshold = 0.05/10150

# Identify significant genes
significant_genes = de_results[de_results['PValue'] < significance_threshold]
num_significant_genes = significant_genes.shape[0]

print(f'Number of significantly differentially expressed genes (P-Value < {significance_threshold}): {num_significant_genes}')

# Optional: Display the significant genes
print(significant_genes.head())
```
